```python
#匹配CSV中的关键字
import re
text = ('In 2017, X company owns the largest market share, \
which is 21 percent more than the second one in 14%. Z company\'s market share is close to that of Y company.\
The remaining share in 55% is left for brands not shown in this chart, including U company, V company, etc.')
print(type('text'))
list_lexicon = ['X company','Y company','V company','more than']
list_keyword = []
for i in range(len(list_lexicon)):
    pattern = re.compile(list_lexicon[i])
    search = re.search(pattern, text)  
    list_keyword.append(search)
print(list_keyword)
```

```python
#提取匹配之后的re.Match object的match字段
print(type(list_keyword))
keywords_matches = []
for i in range(len(list_keyword)):
    keyword_match = str(list_keyword[i])
    tmp = keyword_match.split('\'')[1].split('\'')[0]
    keywords_matches.append(tmp)
print(keywords_matches)
```

```python
#和stanza中的entities进行比较
import ast
print(list_entities)
merge_entities = list(set(list_entities+keywords_matches))
#讲percent 替换成%
merge_entities_tmp = str(merge_entities)
merge_entities = merge_entities_tmp.replace(" percent","%")
#list_merge_entities = json.loads(merge_entities)
merge_entities = ast.literal_eval(merge_entities)
print(type(merge_entities))
print(merge_entities)
```

```python
import sys

input_file = sys.argv[1]
lines = ""
# 读取文件
with open("../examples/assets/data/test.csv", "r",encoding='utf-8') as f:
    lines = f.readlines()
lines = [line.strip() for line in lines]
keys = lines[0].split(',') 
line_num = 1
total_lines = len(lines)
# 数据存储
datas = []
while line_num < total_lines:
        values = lines[line_num].split(",")
        datas.append(dict(zip(keys, values)))
        line_num = line_num + 1
# 序列化时对中文默认使用的ascii编码.想输出真正的中文需要指定ensure_ascii=False
json_str = json.dumps(datas, ensure_ascii=False, indent=4)
# 去除\",\\N,\n 无关符号
result_data = json_str.replace(r'\"','').replace(r'\\N','').replace(r'\n','')
print(result_data)
print(type(result_data))
```
